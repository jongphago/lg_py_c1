### 🔎 key 'students‘의 value로 우리 반 학생들 이름을 입력하여 리스트에 추가하세요.
딕셔너리 변수 myClass에는 위의 표처럼 3가지 아이템이 들어 있습니다.


#### <출력 에시>
```shell
{'name' : DM103 , 'teacher' : 'chloe', 'student' : []}
학생 이름: 주은
학생 이름: 수정
학생 이름: 원준
학생 이름: 지혁
학생 이름: 진영
우리 반 학생들: ['주은', '수정', '원준', '지혁', '진영']
```

````python
# 우리 반 정보를 딕셔너리 변수로 정의해 봅시다.
myClass = {"name": "DM103", "teacher": "", "students": []}

myClass["teacher"] = "레너"
print(myClass)

for i in range(4):
    name = input("학생 이름: ")
    myClass["students"].append(name)

print("우리 반 학생들: {}".format(myClass["students"]))
````

## 마무리 프로젝트 - 뽑기 게임 (2/7)

스스로 마무리 프로젝트를 완성해봅시다.

### 뽑기 게임 규칙
1.  컴퓨터가 1~100사이의 5개의 랜덤한 정수를 정답으로 설정합니다. (이 정수들은 중복 되지 않습니다.)
1. 플레이어가 숫자를 입력합니다. (최대 10번)

1.  게임이 종료되면 정답과 종료 메시지를 출력합니다.

### [게임 성공]
```shell

```
[게임 실패]
```shell
5개의 숫자를 골랐어! 텔레파시를 이용해 숫자를 맞춰봐
---------------------------------
정답 입력: 25
정답 입력: 3
정답 입력: 78
정답 입력: 99
정답 입력: 45

---------------------------------

게임이 종료되었습니다.
```

````python
import random

answers = random.choices(list(range(1, 101)), k=5)
num_try = 0
is_succes = False
print("5개의 숫자를 골랐어! 텔레파시를 이용해 숫자를 맞춰봐")
print("-" * 34)
for _ in range(10):
    guess = int(input("숫자를 맞춰보세요: "))
    if guess in answers:
        print("정답! 게임 성공")
        is_succes = True
        break
if not is_succes:
    print("다시 도전하세요.")
print("-" * 34)
print("정답은: {}".format(answers))
````

````python
import random

numbers = random.choices(list(range(1, 10)), k=3)

for i in range(10):
    print(numbers)
    _guesses = list(input("숫자를 입력하세요.(띄어쓰기 없이 입력 하세요): "))
    guesses = [int(i) for i in _guesses]

    strike = 0
    ball = 0
    out = 0

    for guess, number in zip(guesses, numbers):
        if guess == number:
            strike += 1
        elif guess in numbers:
            ball += 1
        else:
            out += 1
    print(strike, "스트라이크", ball, "볼", out, "아웃")

    if strike == 3:
        print("정답입니다~!")
        break

print("-----------------------------")
print("정답: ", numbers)
print("게임을 종료합니다.")
````
